```python
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd
from requests import get
from bs4 import BeautifulSoup
from warnings import warn
from time import sleep
from random import randint
import numpy as np
```

```python
data = pd.read_csv("casting.csv")
```

```python
data.head()
```

```python
cast_list = data.values.tolist()
```

```python
print(cast_list)
```

```python
nameAC = []
Bdate = []
BirthPlace = []
Linkimg = []
YearsActive = []
infoBox = []
wikiLink= []
for x in cast_list:

  #print(x[0])
  nameAC.append(x[0])
  name = x[0]
  nameWord = name.split()
  #print(nameWord)
  if len(nameWord)<2:
    link=nameWord[0]
  elif len(nameWord)<3:
    link=nameWord[0]+"_"+nameWord[1]
  elif len(nameWord)<4:
    link=nameWord[0]+"_"+nameWord[1]+"-"+(nameWord[2].lower())
  
  print(link)
  url = 'https://en.wikipedia.org/wiki/'+link
  #print(url)
  response = requests.get(url)
  #sleep(randint(8,15))
  if response.status_code == 200:
    wikiLink.append(url)
    soup = BeautifulSoup(response.text, "html.parser")
    DB = None
    for DB in soup.findAll(attrs={'class' : 'bday'}):
      if DB is not None:
        #DB.text.split('-')
        Bdate.append(DB.text[:4])
        print(Bdate)
      else:
        Bdate.append(None)
    bp = None
    for bp in soup.findAll(attrs={'class' : 'birthplace'}):
      if bp is not None:
        BirthPlace.append(bp.text)
      else:
        BirthPlace.append(None)
    imgLinks = []
    for hit in soup.findAll("img"):
        #print(hit['src'])
        imgLinks.append(hit['src'])
    if imgLinks[0] is not None:
      Linkimg.append(imgLinks[0])
    else:
      Linkimg.append(None)
    infoboxData = []
    for bit in soup.findAll(attrs={'class' : 'infobox-data'}):
        #print(hit.text)
        infoboxData.append(bit.text)
    if bit is not None:
      infoBox.append(infoboxData)
    else:
      infoBox.append(None)
    if len(infoboxData)>4:
      if infoboxData[4] is not None:
        YearsActive.append(infoboxData[4])
      else:
        YearsActive.append(None)
    else:
      YearsActive.append(None)
  else:
    Bdate.append(None)
    BirthPlace.append(None)
    Linkimg.append(None)
    YearsActive.append(None)
    infoBox.append(None)
    wikiLink.append(None)
```

```python
print(Bdate)
print(len(infoBox), len(nameAC), len(Bdate), len(BirthPlace),len(Linkimg),len(YearsActive))
```

```python
dat = pd.DataFrame.from_dict({'Ac_Name': nameAC,
                      'Date_Birth': Bdate,
                      'BPlace': BirthPlace,
                      'Img': Linkimg,
                      'Active_Year':YearsActive,
                      'info': infoBox,
                      'WikiLink': wikiLink}, orient='index')
dat = dat.transpose()
dat.fillna(' ',None)
dat.to_csv('CastingData.csv')
```
