```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("movie.csv")
```

```python
data.head()
```

```python
movie_genres=[]
movie_cast=[]
mov_id=[]
movie_genres_temp = data['Genre'].values.tolist()
movie_genres_temp1=[]
```

```python

for i in data['mov_id']:
    movie_genres.append(movie_genres_temp[i-1].split(","))
```

```python
for i in range(len(movie_genres_temp)):
    for j in movie_genres_temp[i].split(","):
        mov_id.append(i)
        movie_genres_temp1.append(j)
    
```

```python
data={"mov_id":mov_id,"movie_genres":movie_genres_temp1}
df = pd.DataFrame(data)
df.head()
```

```python
print(len(pd.unique(df["movie_genres"])))
```

```python
df['genre_id'] = pd.factorize(df['movie_genres'])[0]
```

```python
print(len(pd.unique(df["genre_id"])))
```

```python
df.head()
```

```python
df[['mov_id', 'genre_id']].to_csv('movie_genres.csv')
df[['genre_id','movie_genres']].to_csv('genres.csv')
```
